#### Problem 1
Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.

```python
class Line:
    
    def __init__(self,point_one,point_two):
        self.x_one,self.y_one=point_one
        self.x_two,self.y_two=point_two
        
    def distance(self):
        distance=(((self.x_two-self.x_one)**2+(self.y_two-self.y_one)**2)**0.5)
        print(f"distance is {distance}")
        return distance
        
    def slope(self):
        slope=(self.y_two-self.y_one)/(self.x_two-self.x_one)
        print(f"slope is {slope}")
        return slope
```

```python
# EXAMPLE OUTPUT

coordinate1 = (3,2)
coordinate2 = (8,10)

li = Line(coordinate1,coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

#### Problem 2

Fill in the class 

```python
class Cylinder:

    pi=22/7
    
    def __init__(self,height=1,radius=1):
        self.height=height
        self.radius=radius
        
    def volume(self):
        volume=self.pi*(self.radius**2)*self.height
        print(f"volume is {volume}")
        return volume
        
    def surface_area(self):
        surface_area=2*self.pi*self.radius*(self.radius+self.height)
        print(f"Surface Area is {surface_area}")
        return surface_area
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2,3)
```

```python
c.volume()
```

```python
c.surface_area()
```
