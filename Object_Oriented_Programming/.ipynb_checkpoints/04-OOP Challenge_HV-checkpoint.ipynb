# Object Oriented Programming Challenge

For this challenge, create a bank account class that has two attributes:

* owner
* balance

and two methods:

* deposit
* withdraw

As an added requirement, withdrawals may not exceed the available balance.

Instantiate your class, make several deposits and withdrawals, and test to make sure the account can't be overdrawn.

```python
class Account:

    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance += amount
        print(f"You just added {amount} to your balance and available balance is ${self.balance}")

    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient Balance!")
        else:
            self.balance -= amount
            print(f"You just withdrew {amount} from your account, available balnce is ${self.balance}")

    def __str__(self):
        return f"Account belongs to {self.owner} and available balance is ${self.balance}"

    def __del__(self):
        print(f"Account of {self.owner} succesfully deleted!")
```

```python
# 1. Instantiate the class
acct1 = Account('Jose',100)
```

```python
# 2. Print the object
print(acct1)
```

```python
# 3. Show the account owner attribute
acct1.owner
```

```python
# 4. Show the account balance attribute
acct1.balance
```

```python
# 5. Make a series of deposits and withdrawals
acct1.deposit(5000000)
```

```python
acct1.withdraw(75)
```

```python
# 6. Make a withdrawal that exceeds the available balance
acct1.withdraw(500)
```
