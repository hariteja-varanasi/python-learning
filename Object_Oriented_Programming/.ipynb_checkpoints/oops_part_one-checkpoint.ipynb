```python
class Dog():

    def __init__(self,breed,name,spots):
        self.breed=breed
        self.name=name
        #Expect boolean True/False
        self.spots=spots

    def bark(self):
        print("woof woof!")
```

```python
my_dog=Dog('German Sheperd','Dutta',False)
```

```python
my_dog.breed
```

```python
my_dog.name
```

```python
my_dog.spots
```

```python
my_dog.bark()
```
