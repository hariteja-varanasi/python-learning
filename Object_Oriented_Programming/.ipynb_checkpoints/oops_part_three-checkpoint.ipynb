```python
class Animal():

    def __init__(self):
        print("ANIMAL CREATED")

    def who_am_i(self):
        print("I am an animal")

    def eat(self):
        print("I am eating")
```

```python
my_animal=Animal()
```

```python
class Dog(Animal):

    #CHILD CLASSES NEED NOT IMPLEMENT THE INIT METHOD IF THE PARENT HAS IT IMPLEMENTED
    # def __init__(self):
    #     Animal.__init__(self)
    #     print("Dog Created")

    def who_am_i(self):
        print("I am a dog")

    def eat(self):
        print("I am a Dog and I am eating")

    def bark(self):
        print("WOOF!")
```

```python
my_dog=Dog()
```

```python
my_dog.eat()
```

```python
my_dog.who_am_i()
```

```python
my_dog.bark()
```

```python
my_animal.bark()
```

```python
class Dog():

    def __init__(self,name):
        self.name=name

    def speak(self):
        return self.name + " says WOOF!"
```

```python
class Cat():

    def __init__(self,name):
        self.name=name

    def speak(self):
        return self.name + " says meow!"
```

```python
niko=Dog("niko")
felix=Cat("felix")
```

```python
print(niko.speak())
```

```python
print(felix.speak())
```

```python
for pet in [niko,felix]:
    print(type(pet))
    print(pet.speak())
```

```python
def pet_speak(pet):
    print(pet.speak())
```

```python
pet_speak(niko)
```

```python
pet_speak(felix)
```

```python
class Animal():

    def __init__(self,name):
        self.name=name

    def speak(self):
        raise NotImplementedError("Subclass must implement this abstract method")
```

```python
my_animal=Animal('Fred')
```

```python
my_animal.speak()
```

```python
class Dog(Animal):

    def speak(self):
        return self.name + " says WOOF!"
```

```python
class Cat(Animal):

    def speak(self):
        return self.name + " says meow!"
```

```python
fido=Dog("Fido")
```

```python
isis=Cat("isis")
```

```python
fido.speak()
```

```python
isis.speak()
```
