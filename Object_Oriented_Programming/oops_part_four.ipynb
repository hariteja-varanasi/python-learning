```python
class Book():

    def __init__(self,title,author,pages):
        self.title=title
        self.author=author
        self.pages=pages

    def __str__(self):
        return f"{self.title} by {self.author}"

    def __len__(self):
        return self.pages

    def __del__(self):
        print(f"Book {self.title} by {self.author} has been deleted")
```

```python
b=Book('Python Rocks','Jose',200)
```

```python
print(b)
```

```python
str(b)
```

```python
len(b)
```

```python
del b
```
