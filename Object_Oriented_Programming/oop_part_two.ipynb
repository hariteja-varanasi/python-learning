```python
class Dog():

    #CLASS OBJECT ATTRIBUTE
    #SAME FOR ANY INSTANCE OF THE CLASS
    species='mammal'

    def __init__(self,breed,name,spots):
        self.breed=breed
        self.name=name
        #Expect boolean True/False
        self.spots=False

    #OPERATIONS/ACTIONS ---> Methods
    def bark(self,number):
        #the format parameter should be the self parameter not the constructor input parameter 
        print(f"WOOF! My name is {self.name} and I am {number} years old")
```

```python
my_dog=Dog(breed='German Sheperd',name='Dutta',spots=False)
```

```python
type(my_dog)
```

```python
my_dog.species
```

```python
my_dog.bark(8)
```

```python
class Circle():

    #CLASS OBJECT ATTRIBUTE
    pi=22/7

    def __init__(self,radius=1):
        self.radius=radius
        #the class variable pi can be referred as self.pi or Circle.pi
        self.area=radius*radius*Circle.pi

    #METHOD
    def get_circumference(self):
        return 2*self.pi*self.radius
```

```python
my_circle=Circle(30)
```

```python
my_circle.pi
```

```python
my_circle.radius
```

```python
my_circle.get_circumference()
```

```python
my_circle.area
```
