```python
game_list=[0,1,2]
```

```python
def display_game_list(game_list):
    print("Here is the current list:")
    print(game_list)
```

```python
display_game_list(game_list)
```

```python
def position_choice():
    choice='wrong'
    while choice not in ['0','1','2']:
        choice=input("Pick a position (0,1,2): ")
        if choice not in ['0','1','2']:
            print("Sorry, invalid choice! Please pick (0,1,2)")
    return int(choice)
```

```python
position_choice()
```

```python
def replacement_choice(game_list,position):
    user_input_string=input("Type a string to replace at position: ")
    game_list[position]=user_input_string
    return game_list
```

```python
replacement_choice(game_list,1)
```

```python
def gameon_choice():
    choice='wrong'
    while choice not in ['Y', 'N']:
        choice=input("Do you want to kep playing? (Y or N) ")
        if choice not in ['Y', 'N']:
            print("Sorry, I do not understand. Please choose Y or N")
    if choice=="Y":
        return True
    else:
        return False
```

```python
gameon_choice()
```

```python
game_on=True
game_list=['1','2','3']

while game_on:
    display_game_list(game_list)
    position = position_choice()
    replacement_choice(game_list,position)
    display_game_list(game_list)
    game_on = gameon_choice()
```
